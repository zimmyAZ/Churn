```python
add_nums = DataFrame(merged['msno'], columns=['msno'])
cancelled = DataFrame(merged['msno'], columns=['msno'])
cancelled['is_churn'] = merged['is_churn'].convert_objects(convert_numeric=True)
add_nums['num_25'] = merged['num_25'].convert_objects(convert_numeric=True)
add_nums['num_50'] = merged['num_50'].convert_objects(convert_numeric=True)
add_nums['num_75'] = merged['num_75'].convert_objects(convert_numeric=True)
add_nums['num_985'] = merged['num_985'].convert_objects(convert_numeric=True)
add_nums['num_100'] = merged['num_100'].convert_objects(convert_numeric=True)
add_nums['num_unq'] = merged['num_unq'].convert_objects(convert_numeric=True)
add_nums['total_secs'] = merged['total_secs'].convert_objects(convert_numeric=True)
add_nums['actual_amount_paid'] = merged['actual_amount_paid'].convert_objects(convert_numeric=True)
add_nums['start_year'] = dates['registration_year'].convert_objects(convert_numeric=True)
add_nums['start_month'] = dates['registration_month'].convert_objects(convert_numeric=True)
add_nums['start_day'] = dates['registration_day'].convert_objects(convert_numeric=True)
add_nums['date'] = merged['date'].convert_objects(convert_numeric=True)
add_nums['transaction_date'] = merged['transaction_date'].convert_objects(convert_numeric=True)
add_nums['membership_expire_date'] = merged['membership_expire_date'].convert_objects(convert_numeric=True)
add_nums['registration_init_time'] = merged['registration_init_time'].convert_objects(convert_numeric=True)
added_nums = add_nums.set_index('msno')
cancelled = cancelled.set_index('msno')
```

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()
# add_nums.drop('msno', axis=1, inplace=True)
# cancelled.drop('msno', axis=1, inplace=True)
X = add_nums[0:10000].values
y = cancelled[0:10000].values
x_scaled = min_max_scaler.fit_transform(X)
y_scaled = min_max_scaler.fit_transform(y)
X = x_scaled
y = y_scaled

def sigmoid(x):
    output = 1/(1+np.exp(-x))
    return output

def deriv(x):
    return x*(1-x)

np.random.seed(1)

# syn_best = 11 x 6 x 4 x 2 x 1

syn0 = 2*np.random.random((15,10)) - 1
syn1 = 2*np.random.random((10,10)) - 1
syn2 = 2*np.random.random((10,5)) - 1
syn3 = 2*np.random.random((5,1)) - 1

# alpha_best = 0.01(lowest start) or .001(fast drop)

alpha = .005
epoch = 100000

def train(X, y, syn0, syn1, syn2, syn3, alpha):
    
    for iter in range(epoch):
        
        l0 = X
        l1 = sigmoid(np.dot(l0, syn0))
        l2 = sigmoid(np.dot(l1, syn1))
        l3 = sigmoid(np.dot(l2, syn2))
        l4 = sigmoid(np.dot(l3, syn3))

        l4_error = l4 - y
        l4_delta = l4_error * deriv(l4)
        
        l3_error = l4_delta.dot(syn3.T)
        l3_delta = l3_error * deriv(l3)
        
        l2_error = l3_delta.dot(syn2.T)
        l2_delta = l2_error * deriv(l2)

        l1_error = l2_delta.dot(syn1.T)
        l1_delta = l1_error * deriv(l1)

        syn0 -= alpha * (l0.T.dot(l1_delta))
        syn1 -= alpha * (l1.T.dot(l2_delta))
        syn2 -= alpha * (l2.T.dot(l3_delta))
        syn3 -= alpha * (l3.T.dot(l4_delta))
        
        if(iter % 10000) == 0:
            print('Epoch: ' + str(iter))
            print('Error: ' + str(np.mean(np.abs(l4_error))))
            
    print("Output after training:")
    print(l4)
            
train(X, y, syn0, syn1, syn2, syn3, alpha)

```

```python
weight_1 = syn0
weight_2 = syn1
weight_3 = syn2
weight_4 = syn3
train(X, y, weight_1, weight_2, weight_3,weight_4, alpha)
```
